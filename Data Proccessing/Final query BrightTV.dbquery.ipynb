```sql
select * 
from `workspace`.`default`.`bright_tv_user_profile`;



----JOINING USERPROFILES AND VIEWERSHIPS--
select*
from `workspace`.`default`.`bright_tv_user_profile` AS UP
left join `workspace`.`default`.`bright_tv_viewerships` AS V
ON UP.UserID = V.userid2;



SELECT UP.UserID,
---Dates
V.RecordDate2 as Recording_date,
Dayname(V.RecordDate2) as Day_name,
Monthname(V.RecordDate2) as Month_name,
Dayofmonth(V.RecordDate2) as Day_of_month,

CASE 
When Dayname(V.RecordDate2) = 'Monday' THEN 'Weekday'
When Dayname(V.RecordDate2) = 'Tuesday' THEN 'Weekday'
When Dayname(V.RecordDate2) = 'Wednesday' THEN 'Weekday'
When Dayname(V.RecordDate2) = 'Thursday' THEN 'Weekday'
When Dayname(V.RecordDate2) = 'Friday' THEN 'Weekday'
Else 'Weekend'
End as Day_classification,
Year(V.RecordDate2) as Year,
V.`Duration 2` as Duration,
date_format(V.RecordDate2,'HH:mm:ss') as Purchase_time,

---Times
CASE 
when date_format(V.RecordDate2,'HH:mm:ss') between '00:00:00' and '11:59:59' THEN 'Morning'
when date_format(V.RecordDate2,'HH:mm:ss') between '12:00:00' and '16:59:59' THEN 'Afternoon'
when date_format(V.RecordDate2,'HH:mm:ss') between '17:00:00' and '23:59:59' THEN 'Evening'
end as Time_buckets,

----Age groups

CASE 
WHEN Age < 18 THEN 'Adolescents'
WHEN Age BETWEEN 18 AND 28 THEN 'Young Adults'
WHEN Age BETWEEN 29 AND 39 THEN 'Adult'
WHEN Age BETWEEN 40 AND 54 THEN 'Middle Aged'
WHEN Age BETWEEN 55 AND 64 THEN 'Senior'
WHEN Age >= 65 THEN 'Elderly'
END AS Age_group,

CASE 
WHEN UP.Age < 0 or UP.Age >100 THEN NULL
ELSE UP.Age
END AS Age,

----Cleaning Emails

CASE
WHEN TRIM(UP.Email) LIKE '%@%' THEN TRIM(UP.Email)
ELSE NULL
END AS Email,

------SORTING SOCIAL MEDIA HANDLES

COALESCE(TRIM(UP.`Social Media Handle`),'N/A' ) AS Social_Media_Handles,

----Cleaning channels

V.Channel2,
from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg') AS New_Recording_time,

-----Sorting names
COALESCE(TRIM(UP.Name),'Unknown') as Name,
COALESCE(TRIM(UP.Surname),'Unknown') as Surname,


---GENDERS
CASE 
    WHEN LOWER(TRIM(UP.Gender)) IN('male','m') THEN 'Male'
    WHEN LOWER(TRIM(UP.Gender)) IN('female','f') THEN 'Female'
    ELSE 'Not Specified'
    END AS Gender,

----Allocating race groups
coalesce(TRIM(UP.Race),'Unknown') as Race,

----Province names
Case 
WHEN LOWER(TRIM(UP.Province)) = 'western cape' THEN 'WC'
WHEN LOWER(TRIM(UP.Province)) = 'eastern cape' THEN 'EC'
WHEN LOWER(TRIM(UP.Province)) = 'north west' THEN 'NW'
WHEN LOWER(TRIM(UP.Province)) = 'northern cape' THEN 'NC'
WHEN LOWER(TRIM(UP.Province)) = 'gauteng' THEN 'GP'
WHEN LOWER(TRIM(UP.Province)) = 'kwazulu-natal' THEN 'KZN'
WHEN LOWER(TRIM(UP.Province)) = 'mpumalanga' THEN 'MP'
WHEN LOWER(TRIM(UP.Province)) = 'free state' THEN 'FS'
WHEN LOWER(TRIM(UP.Province)) = 'limpopo' THEN 'LP'
WHEN LOWER(TRIM(UP.Province)) = 'none' THEN 'Unknown'
END AS Province_name


FROM `workspace`.`default`.`bright_tv_user_profile` AS UP
LEFT JOIN `workspace`.`default`.`bright_tv_viewerships` AS V
ON UP.UserID = V.userid2;
```
